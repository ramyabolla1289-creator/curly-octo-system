```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Load and Preprocess Data ---
# Load the IRIS dataset
df = pd.read_csv("/IRIS.csv")

# Create a copy to perform preprocessing
df_processed = df.copy()

# Check for missing values and handle them (IRIS dataset is typically clean, but including for robustness)
for col in df_processed.columns:
    if df_processed[col].isnull().any():
        if pd.api.types.is_numeric_dtype(df_processed[col]):
            df_processed[col] = df_processed[col].fillna(df_processed[col].mean())
        else:
            df_processed[col] = df_processed[col].fillna(df_processed[col].mode()[0])

# Separate numerical and categorical subsets
numerical_cols = df_processed.select_dtypes(include=['number']).columns
categorical_cols = df_processed.select_dtypes(include=['object']).columns

df_numerical = df_processed[numerical_cols]
df_categorical = df_processed[categorical_cols]

# Initialize StandardScaler and fit-transform numerical columns
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df_numerical), columns=numerical_cols, index=df_numerical.index)

# Apply one-hot encoding to categorical columns (e.g., 'species' in IRIS dataset)
df_encoded_categorical = pd.get_dummies(df_categorical, drop_first=True)

# Concatenate the scaled numerical DataFrame and the one-hot encoded categorical DataFrame
df_preprocessed = pd.concat([df_scaled, df_encoded_categorical], axis=1)

print("Data Preprocessing Complete.")
print("First 5 rows of preprocessed DataFrame:")
print(df_preprocessed.head())
print("\nInformation about the preprocessed DataFrame:")
df_preprocessed.info()

# --- 2. Determine Optimal Clusters (Elbow Method) ---
wcss = []
for i in range(1, 11):
    kmeans_elbow = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans_elbow.fit(df_preprocessed)
    wcss.append(kmeans_elbow.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
print("\nElbow Method plot generated to determine optimal number of clusters.")

# --- 3. Apply K-Means Clustering (using n_clusters=3 based on Elbow Method) ---
optimal_clusters = 3
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', max_iter=300, n_init=10, random_state=42)
kmeans.fit(df_preprocessed)
df_preprocessed['cluster'] = kmeans.labels_

print(f"\nK-Means clustering applied with {optimal_clusters} clusters. Cluster labels added.")
print("First 5 rows of df_preprocessed with cluster labels:")
print(df_preprocessed.head())

# --- 4. Visualize Clusters with PCA ---
# Exclude the 'cluster' column for PCA fitting
X_for_pca = df_preprocessed.drop(columns=['cluster'])

# Instantiate PCA with 2 components
pca = PCA(n_components=2, random_state=42)

# Apply PCA to reduce the data to two principal components
pca_components = pca.fit_transform(X_for_pca)

# Create a new DataFrame for the PCA results, including the two principal components and the 'cluster' labels
df_pca = pd.DataFrame(data=pca_components, columns=['PC1', 'PC2'])
df_pca['cluster'] = df_preprocessed['cluster']

# Create a scatter plot of the two principal components, with points colored by their assigned cluster
plt.figure(figsize=(10, 8))
sns.scatterplot(x='PC1', y='PC2', hue='cluster', data=df_pca, palette='viridis', s=100, alpha=0.7)
plt.title('Clusters Visualized with PCA (2 Components)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.legend(title='Cluster')
plt.show()
print("\nClusters visualized using PCA.")

# --- 5. Analyze Cluster Characteristics ---
# Combine original features with cluster labels for interpretability
df_original_features_processed = df.copy()

# Handle missing values (if any) - as per previous preprocessing
for col in df_original_features_processed.columns:
    if df_original_features_processed[col].isnull().any():
        if pd.api.types.is_numeric_dtype(df_original_features_processed[col]):
            df_original_features_processed[col] = df_original_features_processed[col].fillna(df_original_features_processed[col].mean())
        else:
            df_original_features_processed[col] = df_original_features_processed[col].fillna(df_original_features_processed[col].mode()[0])

df_original_features_processed['cluster'] = df_preprocessed['cluster']

# Analyze numerical features by cluster
cluster_summary_numerical = df_original_features_processed.groupby('cluster')[numerical_cols].mean()

# Analyze categorical features by cluster (showing counts or proportions)
cluster_summary_categorical = pd.crosstab(df_original_features_processed['cluster'], df_original_features_processed['species'])

print("\nNumerical Feature Means by Cluster:")
print(cluster_summary_numerical)

print("\nCategorical Feature Counts by Cluster (Species):")
print(cluster_summary_categorical)
print("\nCluster characteristics analysis complete.")
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

df = pd.read_csv("/AI_Resume_Screening.csv")

job_description = """
Looking for a Data Analyst with Python, SQL,
Machine Learning, and Data Visualization skills.
"""
model = SentenceTransformer('all-MiniLM-L6-v2')

resume_texts = df['Skills'].astype(str).tolist()
resume_embeddings = model.encode(resume_texts)
job_embedding = model.encode([job_description])

similarity_scores = cosine_similarity(resume_embeddings, job_embedding)
df['Similarity_Score'] = similarity_scores

threshold = 0.50

df_filtered = df[df['Similarity_Score'].apply(lambda x: x > threshold)]

df_filtered = df_filtered.sort_values(by='Similarity_Score', ascending=False)

required_skills = ["python", "sql", "machine learning",
                   "data visualization", "excel", "nlp"]

def skill_match(text):
    text = text.lower()
    matched = [skill for skill in required_skills if skill in text]
    missing = [skill for skill in required_skills if skill not in text]
    return matched, missing

df_filtered['Matched_Skills'] = df_filtered['Skills'].apply(lambda x: skill_match(str(x))[0])
df_filtered['Missing_Skills'] = df_filtered['Skills'].apply(lambda x: skill_match(str(x))[1])

print(f"\n Candidates with Score > {threshold}\n")

if len(df_filtered) == 0:
    print("No candidates met the threshold.")
else:
    for index, row in df_filtered.iterrows():
        print(f"Candidate: {row['Name']}")
        print(f"Score: {row['Similarity_Score']:.3f}")
        print(f"Matched Skills: {row['Matched_Skills']}")
        print(f"Missing Skills: {row['Missing_Skills']}")
        print("-" * 40)
```
