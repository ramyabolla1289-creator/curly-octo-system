```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor



df=pd.read_csv("/content/creditcard.csv.zip")

np.random.seed(42)

n_users = 50
n_transactions = 1000

users = np.random.randint(1, n_users + 1, n_transactions)

dates = pd.date_range(start="2023-01-01", periods=n_transactions, freq="H")

amounts = np.random.normal(loc=2000, scale=500, size=n_transactions)


anomaly_indices = np.random.choice(n_transactions, 20, replace=False)
amounts[anomaly_indices] = amounts[anomaly_indices] * 8

df = pd.DataFrame({
    "user_id": users,
    "transaction_date": dates,
    "amount": amounts
})

print("Dataset Shape:", df.shape)



df['transaction_date'] = pd.to_datetime(df['transaction_date'])

df['hour'] = df['transaction_date'].dt.hour
df['day_of_week'] = df['transaction_date'].dt.dayofweek

df['amount_normalized'] = df.groupby('user_id')['amount'] \
                             .transform(lambda x: (x - x.mean()) / x.std())

df['amount_normalized'].fillna(0, inplace=True)



features = df[['amount_normalized', 'hour', 'day_of_week']]

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)



iso = IsolationForest(contamination=0.02, random_state=42)
df['anomaly_iforest'] = iso.fit_predict(scaled_features)

df['anomaly_iforest'] = df['anomaly_iforest'].map({1: 0, -1: 1})



lof = LocalOutlierFactor(n_neighbors=20, contamination=0.02)
df['anomaly_lof'] = lof.fit_predict(scaled_features)

df['anomaly_lof'] = df['anomaly_lof'].map({1: 0, -1: 1})


plt.figure(figsize=(12,6))

normal = df[df['anomaly_iforest'] == 0]
plt.scatter(normal['transaction_date'], normal['amount'], alpha=0.4)

anomaly = df[df['anomaly_iforest'] == 1]
plt.scatter(anomaly['transaction_date'], anomaly['amount'])

plt.title("Anomalous Transactions Over Time (Isolation Forest)")
plt.xlabel("Transaction Date")
plt.ylabel("Transaction Amount")
plt.show()


anomalies = df[df['anomaly_iforest'] == 1]

print("\nNumber of Anomalies Detected:", anomalies.shape[0])

print("\nSample Flagged Transactions:")
print(anomalies[['user_id',
                 'transaction_date',
                 'amount',
                 'amount_normalized',
                 'hour',
                 'day_of_week']].head())



def explain_transaction(row):
    reasons = []

    if abs(row['amount_normalized']) > 3:
        reasons.append("Unusually high compared to user's normal spending")

    if row['hour'] < 5:
        reasons.append("Transaction at unusual late-night hour")

    if row['day_of_week'] in [5, 6]:
        reasons.append("Weekend transaction pattern deviation")

    if len(reasons) == 0:
        reasons.append("Isolated behavior detected by model pattern")

    return "; ".join(reasons)


anomalies['explanation'] = anomalies.apply(explain_transaction, axis=1)

print("\nDetailed Explanation for First 5 Anomalies:")
print(anomalies[['user_id',
                 'transaction_date',
                 'amount',
                 'explanation']].head())


df.to_csv("anomaly_detection_results.csv", index=False)

print("\nResults saved to anomaly_detection_results.csv")
```
